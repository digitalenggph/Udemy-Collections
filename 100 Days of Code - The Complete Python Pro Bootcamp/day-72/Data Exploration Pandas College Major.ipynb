## Sorting Values & Adding Columns: 
### Majors with the Most Potential vs Lowest Risk


```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

## Top 5 degrees with highest potential
#### (highest values in the 90th percentile)

```python
high_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

## Top 5 degrees with highest risk
#### (difference between the 10th percentile and the 90th percentile)

```python
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

## Top 5 degrees with highest Mid-Career Median Salary
#### (highest values in their Mid-Career Median Salary)

```python
highest_mid_career_salary = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
highest_mid_career_salary[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```
